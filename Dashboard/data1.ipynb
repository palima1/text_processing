```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import re
```

```python
df = pd.read_csv('data_final.csv')
df.head()
```

```python
df.info()
```

```python
def lab_valid(score):
    if score <= 2:
        return "Ruim"
    elif score == 3:
        return "Neutro"
    else:
        return "Bom"
```

```python
df['sentimento'] = df.score.apply(lab_valid)

df.head(30)
```

```python
df.to_csv('data_sentimento.csv', index=False)
```
