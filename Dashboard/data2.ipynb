```python
import pandas as pd
import re
```

```python
df = pd.read_csv("mouse.csv")
df.head()
```

```python
df.info()
```

```python
mes = {'janeiro': '01', 'fevereiro': '02', 'março': '03', 'abril': '04', 'maio': '05', 'junho': '06', 'julho': '07', 
         'agosto': '08', 'setembro': '09', 'outubro': '10', 'novembro': '11', 'dezembro': '12'}
```

```python
def prep(x):
  x['score'] = x['score'].replace(' de 5 estrelas', '')
  x['score'] = x['score'].replace(',', '.')
  
  x['description'] = str(x['description']).replace('\n  ', '')
  x['description'] = str(x['description']).replace('\n', '')

  r = re.search('de \w* de', x['date']).span()
  x['date'] = x['date'].replace(x['date'][r[0]-1:r[1]+1], '/' + mes[x['date'][r[0]+3 : r[1]-3]] + '/')

  return x
```

```python
df.apply(prep, axis=1)
```

```python
df.to_csv('data_final.csv', index=False)
```
