```python
import pandas as pd
import plotly.express as px
#import streamlit as st
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('data_sentimento.csv')
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], dayfirst=True)
```

```python
df.info()
```

```python
def get_mes(date):
    return date.strftime("%b")

df['mes'] = df.date.apply(get_mes)
df.head()
```

```python
def get_ano(date):
    return date.strftime("%y")

df['ano'] = df.date.apply(get_ano)
df.head()
```

```python
cont = pd.DataFrame(df.groupby(['mes', 'ano', 'sentimento'])['sentimento'].count())
cont.columns = ['cont']
cont.head()
```

```python
cont = cont.reset_index()
cont.head()
```

```python
nova_data = []

for row in cont.itertuples():
    nova_data.append(f'{row.mes}-{row.ano}')

cont['novas_data'] = nova_data
cont.head()
```

```python
cont['ano'] = pd.to_numeric(cont['ano'])
```

```python
cont_filtro = cont.query("ano == 21 or ano == 22")
cont_filtro.head()
```

```python
cont_filtro['novo_form'] = pd.to_datetime(cont_filtro['novas_data'], format='%b-%y')
cont_filtro.head()
```

```python
cont_filtro = cont_filtro.sort_values('novo_form')
```

```python
cont_filtro.head()
```

```python
cont_filtro.novas_data.unique()
```

```python
percentages = []

for dif_mes_ano in cont_filtro.novas_data.unique():
    print(dif_mes_ano)
    dataset = cont_filtro.query(f'novas_data == "{dif_mes_ano}"')
    total = sum(dataset['cont']) 
    print(total)
    
    mes_ano_values = dataset['cont'].values
    
    if mes_ano_values.size > 0:
        for value in mes_ano_values:
            percentage = (value * 100)/total
            print(percentage)
            percentages.append(percentage)
    else:
        pass
        
cont_filtro['mes_perc'] = percentages 
```

```python
cont_filtro.head(20)
```

```python
cont_filtro.to_csv('mes_cont.csv', index=False)
```
