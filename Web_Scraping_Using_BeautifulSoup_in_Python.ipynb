```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
link="https://www.amazon.com.br/dp/B075XMBJXJ?ie=UTF8&linkCode=sl1&tag=guiaesperto04-20&linkId=b74c40d06f45c681fc04a0785bf2ff0d&language=pt_BR&ref_=as_li_ss_tl"
```

```python
page=requests.get(link)
print(page.content)
```

```python
a= bs (page.content, 'html.parser')

review_names=a.find_all('span',class_='a-profile-name')
print(review_names)
```

```python
cust_name=[]
for i in range(0, len(review_names)):
  cust_name.append(review_names[i].get_text())
  print(cust_name.pop(0))
```

```python
title=a.find_all('a',class_='review-title-content')
print(title)
```

```python
review_title=[]
for i in range(0, len(title)):
  review_title.append(title[i].get_text())
print(review_title)

review_title[:]=[titles.lstrip('\n')for titles in review_title]
print(review_title)

review_title[:]=[titles.rstrip('\n')for titles in review_title]
print(review_title)
```

```python
rating=a.find_all('i',class_='review-rating')
print(rating)
```

```python
rate=[]
for i in range (0, len(rating)):
  rate.append(rating[i].get_text())
print(rate)
```

```python
review=a.find_all('span',{'data-hook':'review-body'})
print(review)
```

```python
review_content=[]
for i in range(0,len(review)):
  review_content.append(review[i].get_text())
print(review_content)
```

```python
review_content[:]=[reviews.lstrip('\n')for reviews in review_content]
print(review_content)
```

```python
review_content[:]=[reviews.rstrip('\n')for reviews in review_content]
print(review_content)
```

```python
data=a.find_all('span',{'data-hook':'review-date'})
data
```
