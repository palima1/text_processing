```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

reviewlist = []

def get_soup(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    return soup


def get_reviews(soup):
    reviews = soup.find_all('div', {'data-hook': 'review'})
    try:
        for item in reviews:
            review = {
            'score': item.find('i', {'data-hook': 'review-star-rating'}).text.replace(' de 5 estrelas', '').strip(),
            'description': item.find('span', {'data-hook': 'review-body'}).text.strip(),
            'date': item.find('span',{'data-hook':'review-date'}).text.replace('Avaliado no Brasil em ','').strip(),
            }
            reviewlist.append(review)
    except:
        pass

for x in range(1,1200):
    soup = get_soup(f'https://www.amazon.com.br/Cmera-Monitoramento-Tapo-C200-TP-Link/product-reviews/B07XLML2YS/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={x}')
    print(f'Getting page: {x}')
    get_reviews(soup)
    print(len(reviewlist))
    if not soup.find('li', {'class': 'a-disabled a-last'}):
        pass
    else:
        break

df = pd.DataFrame(reviewlist)
df.to_csv('camera_data.csv', index=False)
print('Fin.')
```
